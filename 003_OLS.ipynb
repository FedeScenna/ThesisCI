```python
import pandas as pd
import numpy as np
from utils import w_data
import statsmodels.api as sm
```

```python
df = pd.read_csv(w_data)
df.head()
```

## Model 1: 
Extremely simple model. 
Target variable: Weekly_Sales
Explanatory: 

```python
model1_df = df.groupby("Date").agg({"Weekly_Sales":"sum", "Fuel_Price":"max"}).reset_index()
X, Y = np.log(model1_df["Fuel_Price"]),np.log(model1_df["Weekly_Sales"])
model_simple = sm.OLS(Y, X)
results = model_simple.fit()
print(results.summary())
```

## Model 2: 
same as model 1 but store type as dummies

```python
model2_df = df.groupby(["Date", "Type", "IsHoliday"]).agg({"Weekly_Sales":"sum", "Fuel_Price":"max"}).reset_index()
model2_df = pd.concat([model2_df, pd.get_dummies(model2_df["Type"])], axis = 1)
model2_df
#X, Y = np.log(model1_df["Fuel_Price"]),np.log(model1_df["Weekly_Sales"])
#model_simple = sm.OLS(Y, X)
#results = model_simple.fit()
#print(results.summary())
```
